```python
import tensorflow as tf
import pickle
import pandas as pd
import numpy as np
from tensorflow.keras.models import load_model
```

```python
# Loading Model from pickle file
model = load_model("C:/Users/gaurv/Downloads/STUDY/Deep Learning/Implementations/ANN/Models/model.h5")

# Loading scaler and encoders from pickle file
with open("C:/Users/gaurv/Downloads/STUDY/Deep Learning/Implementations/ANN/Models/scaler.pkl","rb") as file:
    scaler = pickle.load(file)

with open("C:/Users/gaurv/Downloads/STUDY/Deep Learning/Implementations/ANN/Models/one_hot_encoder.pkl","rb") as file:
    one_hot_encoder = pickle.load(file)

with open("C:/Users/gaurv/Downloads/STUDY/Deep Learning/Implementations/ANN/Models/label_encoder.pkl","rb") as file:
    label_encoder = pickle.load(file)
```

```python
input_data = {
'CreditScore': 600,
'Geography': 'France',
'Gender': 'Male',
'Age': 40,
'Tenure': 3,
'Balance': 60000,
'NumOfProducts': 2,
'HasCrCard': 1,
'IsActiveMember': 1,
'EstimatedSalary': 50000
}
```

```python
geography_encoded = one_hot_encoder.transform([[input_data['Geography']]]).toarray()
```

```python
geography_encoded = pd.DataFrame(geography_encoded,columns=one_hot_encoder.get_feature_names_out())
geography_encoded
```

```python
input_data = pd.DataFrame([input_data])
input_data
```

```python
input_data = pd.concat([input_data.reset_index(drop=True),geography_encoded],axis=1)
input_data
```

```python
input_data.drop('Geography',axis=1,inplace=True)
```

```python
input_data['Gender'] = label_encoder.transform(input_data['Gender'])
input_data
```

```python
input_data_scaled = scaler.transform(input_data)
input_data_scaled
```

```python
prediction = model.predict(input_data_scaled)
prediction
```

```python
prediction_proba = prediction[0][0]
prediction_proba
```

```python
if prediction_proba > 0.5:
    print("Customer is likely to churn")
else:
    print("Customer is not likely to churn")
```
